```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import pandas as pd
import seaborn as sns
```

```python
df=pd.read_csv('C:/Users/Srinivas/Downloads/Desktop/testdata/Test_Data.csv')
```

```python
datasets = {}
by_class = df[['A','B','C','G','R','X6','X8','Class']].groupby(['Class'])
df.Class.value_counts()
for groups, data in by_class:
    datasets[groups] = data
```

```python

aclass=datasets['A']
bclass=datasets['B']
cclass=datasets['C']
dclass=datasets['D']
eclass=datasets['E']
fclass=datasets['F']
```

```python
dfa=pd.read_csv('C:/Users/Srinivas/Downloads/Desktop/testdata/aclass.csv')
dfa=dfa.drop(['Unnamed: 0'],axis=1)
dfa.info()
```
